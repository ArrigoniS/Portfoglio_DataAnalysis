# From website to .CSV

## Let's start

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

### Url of the website

```python
url = 'https://en.wikipedia.org/wiki/List_of_the_largest_software_companies'
page = requests.get(url)
```

### Finding the right data

```python
soup = BeautifulSoup(page.text, 'html') # import the html code of the website
table = soup.find('table', class_ = 'wikitable sortable') # find the right table, we can also use find_all
```

```python
# Bring title in
Titles = table.find_all('th') 
List_of_titles = [title.text.strip() for title in Titles] 
print(List_of_titles) #check
```

## Import the data in pandas' dataframe

```python
df = pd.DataFrame(columns = List_of_titles) #Title

                                            # DATA
    
Data_table = table.find_all('tr')[1:] 

# loop in every row to create a single list for every row and add to our dataframe

for row in Data_table:
    row_data = row.find_all('td')
    list_of_single_row = [data.text.strip() for data in row_data]
    
    length = len(df)
    df.loc[length] = list_of_single_row
```

```python
#Check if everything it'ok    
df
```

### Create the file .csv

```python
path = input('Insert the path of the folder that you want to put in the data: ')
file_name = input('To export the data, you need to create a file, choose a name: ')
df.to_csv(path + '/' + file_name + '.csv', index = False)
```
